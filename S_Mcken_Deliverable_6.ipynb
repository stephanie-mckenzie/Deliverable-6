```python
# Stephanie McKenzie
# ISM4300 - Deliverable (6)
# February 23, 2023


# Create data frames
import pandas as pd
import numpy as mp
import seaborn as sns

# Load data set
iris = sns.load_dataset('iris')

# Drop the species column
del iris["species"]
iris.head()

```

```python
# Add columns
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
# Make summary table

mean = list(iris.mean())
stdev = list(iris.std())
mini = list(iris.min())
maxi = list(iris.max())


summary_dict = {'Mean':mean, 'Standard Deviation': stdev,'Minimum': mini, 'Maximum': maxi }

summary_table = pd.DataFrame(summary_dict)

summary_table.head()


```
